```python
import pandas as pd
```

```python
df = pd.read_excel("Data/population_total.xlsx")
```

```python
df
```

```python
df_melted = pd.melt(df, id_vars=['country'], value_vars=[ counter for counter in range(1800, 2101) ],
                    var_name='year', value_name='population')
```

```python
df_melted
```

```python
null_count = df_melted.isnull().sum()
print("Null values in each column:")
print(null_count)
```

```python
print(df_melted.dtypes)
df_melted = df_melted.astype({
    'country': 'string',  
    'year': 'int32'       
})
print(df_melted.dtypes)
```

```python
def transform_population(population):
    if pd.isna(population):
        return None
    elif isinstance(population, str) :
        if population[-1] == 'B':
            return float(population[:-1]) * 1000000000
        if population[-1] == 'M':
            return float(population[:-1]) * 1000000
        elif population[-1] == 'k':
            return float(population[:-1]) * 1000
        else :
            return float(population)
    elif isinstance(population, (int,float)) :
        return float(population)  
    else :
        return None
    
df_melted['population']=df_melted['population'].apply(transform_population)
df_melted
```

```python
df_melted.to_csv("Data/population_total.csv", index=False, header=True , sep=';' )
```
