```python
import pandas as pd
from selenium import webdriver
```

```python
fathi = webdriver.Chrome()
```

```python
fathi.get("https://www.boxofficemojo.com/chart/top_lifetime_gross/?area=XWW")
```

```python
full_elements = fathi.find_elements("xpath", "//td")
print(len(full_elements))
```

```python
for element in full_elements: 
    print(element.text)
```

```python
data = []
for element in range(0, 800 ,4):
    Rank           = full_elements[element].text
    Title          = full_elements[element + 1].text
    Lifetime_Gross = full_elements[element + 2].text
    Year           = full_elements[element + 3].text
    data.append([Rank, Title, Lifetime_Gross, Year])
```

```python
df = pd.DataFrame(data)
df.to_csv("top_200_movies.csv", header=["Rank", "Title", "Lifetime Gross", "Year"] , index=False)
fathi.quit()
```
