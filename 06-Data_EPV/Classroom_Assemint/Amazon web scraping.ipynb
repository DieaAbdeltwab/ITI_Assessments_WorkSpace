```python
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException
```

```python
driver = webdriver.Chrome()
```

```python
driver.get("https://www.amazon.eg/s?k=samsung&rh=p_89%3Asamsung&language=en&ref=SQEG-MEB-SR301")
```

```python
allproducts = driver.find_elements("xpath", "//div[@class='a-section a-spacing-base']")
```

```python
for product in allproducts:
    try:
        title = product.find_element("xpath", ".//h2[@class='a-size-base-plus a-spacing-none a-color-base a-text-normal']/span").text
    except NoSuchElementException:
        title = "N/A"

    try:
        price = product.find_element("xpath", ".//span[@class='a-price-whole']").text
    except NoSuchElementException:
        price = "N/A"

    try:
        imgurl = product.find_element("xpath", ".//img[@class='s-image']").get_attribute("src")
    except NoSuchElementException:
        imgurl = "N/A"

    print(f"Title: {title}")
    print(f"Price: {price}")
    print(f"Image URL: {imgurl}")
    print("-" * 100)  
```
