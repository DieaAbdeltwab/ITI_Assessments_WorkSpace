## **Introduction to Seaborn**

### What is Seaborn?

> - Seaborn is a Python data visualization library built on **top of Matplotlib**. It provides a high-level interface for creating attractive and informative statistical graphics. Seaborn is particularly well-suited for working with **Pandas DataFrames** and makes it easy to visualize complex datasets.
> - Its **dataset-oriented**, declarative API lets you focus on what the different elements of your plots mean, rather than on the details of how to draw them.

```python
# Import seaborn
import seaborn as sns
import matplotlib.pyplot as plt
```

## **Basic Plots**

### **1. Line Plots (`sns.lineplot()`)**

```python
# Load a dataset
flights = sns.load_dataset("flights")
flights.head(10)
```

```python
# Line plot
plt.figure(figsize=(10, 6))

sns.lineplot(x="year", y="passengers", data=flights, errorbar=None) # errorbar=None
# sns.lineplot(x=flights["year"], y=flights["passengers"])

plt.title("Passengers Over Time")
plt.xlabel("Year")
plt.ylabel("Passengers")
```

### **2. Scatter Plots (`sns.scatterplot()`)**

```python
# Load a dataset
tips = sns.load_dataset("tips")
tips.head()
```

```python
# Scatter plot
plt.figure(figsize=(8, 6))

sns.scatterplot(x="total_bill", y="tip", data=tips, hue='smoker', palette='Set1') # hue='time' palette='Set1'

plt.title("Total Bill vs Tip")
plt.xlabel("Total Bill ($)")
plt.ylabel("Tip ($)")
```

### **3. Bar Plots (`sns.barplot()`)**

```python
# Bar plot
plt.figure(figsize=(8, 6))

sns.barplot(x="day", y="total_bill", data=tips, errorbar=None, hue='sex', palette='Set1') # hue='smoker', palette='Set1'

plt.title("Average Total Bill by Day")
plt.xlabel("Day")
plt.ylabel("Average Total Bill ($)")
plt.show()
```

```python
# Bar plot
plt.figure(figsize=(8, 6))

sns.barplot(x="day", y="total_bill", data=tips, hue='sex', palette='Set2', errorbar=None)
plt.title("Average Total Bill by Day")
plt.xlabel("Day")
plt.ylabel("Average Total Bill ($)")
```

### **4. Count Plots (`sns.countplot()`)**

> Show the counts of observations in each categorical bin using bars. A count plot can be thought of as a histogram across a categorical, instead of quantitative, variable.

```python
titanic = sns.load_dataset("titanic")
titanic.head()
```

```python
sns.countplot(data=titanic, x="class", color='m')
```

```python
sns.countplot(data=titanic, x="class", hue="survived", palette='Set1')
```

```python
sns.countplot(data=titanic, y="class", hue="survived", palette='Set1')

plt.legend()
```

### **5. Histograms (`sns.histplot()`)**

```python
tips.head()
```

```python
# Histogram
plt.figure(figsize=(8, 6))

sns.histplot(x=tips["total_bill"], bins=10) # color='blue'

plt.title("Distribution of Total Bill")
plt.xlabel("Total Bill ($)")
plt.ylabel("Frequency")
plt.show()
```

### **6. Kernel Density Estimates (`sns.kdeplot()`)**

```python
# KDE plot
plt.figure(figsize=(8, 6))

sns.kdeplot(tips["total_bill"], fill=True, color="green")

plt.title("Kernel Density Estimate of Total Bill")
plt.xlabel("Total Bill ($)")
plt.ylabel("Density")
plt.show()
```

### **7. Box Plot (`sns.boxplot()`)**

```python
# Box plot
plt.figure(figsize=(8, 6))

sns.boxplot(x="day", y="total_bill", data=tips, palette="Set2")

plt.title("Total Bill by Day")
plt.xlabel("Day")
plt.ylabel("Total Bill ($)")
plt.show()
```

### **8. Pair Plot (`sns.pairplot()`)**

> - Plot pairwise relationships in a dataset.
> - By default, this function will create a grid of Axes such that each numeric variable in data will by shared across the y-axes across a single row and the x-axes across a single column. The diagonal plots are treated differently: a **univariate** distribution plot is drawn to show the marginal distribution of the data in each column.
> - The simplest invocation uses `scatterplot()` for each pairing of the variables and `histplot()` for the marginal plots along the diagonal:

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
sns.pairplot(iris)
```

Assigning a `hue` variable adds a semantic mapping and changes the default marginal plot to a layered kernel density estimate (KDE):

```python
sns.pairplot(iris, hue='species') # diag_kind="hist"
```

The `markers` parameter applies a style mapping on the off-diagonal axes. Currently, it will be redundant with the `hue` variable:

```python
sns.pairplot(iris, hue='species', markers=["*", "o", ">"])
```

Use `vars` or `x_vars` and `y_vars` to select the variables to plot:

```python
# sns.pairplot(iris, hue='species', x_vars=['sepal_width', 'petal_width'], y_vars=['sepal_width', 'petal_width'])
sns.pairplot(iris, hue='species', vars=['sepal_length', 'petal_width'])
```

### **9. Heatmap (`sns.heatmap()`)**

```python
flights = sns.load_dataset('flights')
flights.head()
```

```python
new_flights = flights.pivot(index='month', columns='year', values='passengers')
new_flights
```

```python
sns.heatmap(new_flights)
```

```python
sns.heatmap(new_flights, annot=True, fmt='d')
```

```python
sns.heatmap(new_flights, annot=True, fmt='d', cmap='crest')
```

#### **Correlation Matrix**

```python
iris_corr = iris.corr(numeric_only=True)
iris_corr
```

```python
sns.heatmap(iris_corr, annot=True, cmap='GnBu')
```
